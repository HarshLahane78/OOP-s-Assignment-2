```python
"""To prevent a class from having multiple instances in Python, you can use a design pattern called the Singleton Pattern. 
The Singleton Pattern ensures that a class has only one instance and provides a global point of access to that instance. 
Typically, this is achieved by controlling the instantiation process, allowing only one instance 
to be created and providing a way to access it."""

class SingletonClass:
    
    _instance = None

    def __new__(cls):
        
        if not cls._instance:
            
            cls._instance = super(SingletonClass, cls).__new__(cls)
        return cls._instance

    def __init__(self, data):
        
        if not hasattr(self, 'initialized'):
            self.data = data
            self.initialized = True

            
ins1 = SingletonClass.__new__(SingletonClass)
ins1.__init__(data="instance data 1")

ins2 = SingletonClass.__new__(SingletonClass)
ins2.__init__(data="instance data 2")

print("Instance : ",ins1 is ins2)  


print(ins1.data)  
print(ins2.data) 
```

```python
"""you can access the methods and attributes of a parent class from a child class by using the super() function. 
The super() function returns a temporary object of the superclass, allowing you to call its methods and access its attributes. 
This is particularly useful when you want to extend or override methods in the child class while still 
utilizing the functionality provided by the parent class."""

class animal:
    
    def __init__(self,species):
        self.species = species
        
    def speak(self):
        return f"the species of the animal is : {self.species}"
    
class elephant(animal):
    
    def __init__(self,species,breed):
        super().__init__(species)
        self.breed = breed
        
    def speak(self):
        return f"The breed of {self.species} is : {self.breed}"
    
    
elp1 = elephant("elephant","african")

print("the species of the animal :",elp1.species)
print(elp1.speak())
```

```python
"""Abstract Base Classes (ABCs) in Python provide a way to define a common interface for a group of related classes.
They allow you to define abstract methods that must be implemented by the concrete (sub)classes that inherit from them. 
ABCs are part of the abc module in Python."""

from abc import ABC , abstractmethod

class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass
    
class Circle(Shape):
    
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        return math.pi * self.radius**2
    
class Rectangle(Shape):
    
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
        
    def area(self):
        return self.length*self.breadth
    
    
cr1 = Circle(7)
rect1 = Rectangle(12,9)

print("the area of the circle is : ",cr1.area())
print("the area of the rectangle is : ",rect1.area())
```

```python
class BankAccount :
    
    def __init__(self):
        self.__balance = 1000
        self.__account_no = 12345
        
    def account_details(self):
        print("Account Details : ")
        print(f"Balance : {self.__balance}")
        print(f"Account Number : {self.__account_no}\n")
        
    def deposite(self,deposite_money):
        
        if deposite_money > 0 :
            self.__balance += deposite_money
            print(f"Money Deposited Successfully. Your Current Balance : {self.__balance}")
        
        else :
            print("Invalid input . Please enter the valid input ")
            
    def withdraw(self,withdraw_money):
        
        if withdraw_money <= self.__balance:
            self.__balance -= withdraw_money
            print(f"Money Withdrawed Successfully. Your Current Balance : {self.__balance}")
            
        else :
            print("You don't have enough money")
        

acc1 = BankAccount()     
acc1.account_details()
acc1.deposite(1000)
acc1.withdraw(500)
```

```python
"""Q. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators
like `+` and `*`.

Operator overloading in Python allows you to define how operators behave for user-defined objects. By overloading operators, 
you can specify the behavior of standard Python operators such as +, *, ==, and others for instances of your custom classes. 
This concept is closely related to polymorphism, as it enables objects of different types to be used with the same operators in
a way that makes sense for each type.

Relationship to Polymorphism
Operator overloading is a form of polymorphism in Python. It allows objects of different types to be used with the same operator,
and the behavior of the operator depends on the type of the objects involved. The same operator can behave differently for different types, 
providing a common interface for working with diverse objects."""

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Point):
    
            return Point(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            
            return Point(self.x + other, self.y + other)
        else:
        
            raise TypeError(f"Unsupported operand type: {type(other)}")

    def __mul__(self, scalar):
        
        if isinstance(scalar, (int, float)):
            return Point(self.x * scalar, self.y * scalar)
        else:
        
            raise TypeError(f"Unsupported operand type: {type(scalar)}")


point1 = Point(1, 2)
point2 = Point(3, 4)

result_add = point1 + point2
print(f"Addition Result: {result_add}")

result_add_scalar = point1 + 5
print(f"Addition with Scalar Result: {result_add_scalar}")

result_multiply = point1 * 2
print(f"Multiplication Result: {result_multiply}")
```

```python
"""Encapsulation is a fundamental concept in object-oriented programming that involves bundling data (attributes) 
and methods (functions) that operate on that data into a single unit, typically a class. It helps in hiding the internal 
implementation details of a class and exposing only what is necessary for the outside world. Encapsulation plays a crucial 
role in achieving abstraction, and here's why:

Significance of Encapsulation in Achieving Abstraction:

Information Hiding:

Encapsulation allows you to hide the internal details of a class from the outside world. This means that the implementation details,
such as data members and specific methods, are not directly accessible to external code.

Reduced Complexity:

By encapsulating the implementation details within a class, the external code interacts with a simplified interface. 
This reduces the complexity seen by users of the class, making it easier to understand and use.

Modularity:

Encapsulation promotes modularity by organizing code into self-contained units (classes). Each class encapsulates a specific 
set of functionalities, and changes made within a class have minimal impact on the rest of the codebase."""

class BankAccount :
    
    def __init__(self):
        self._balance = 1000
        self._account_no = 12345
        
    def account_details(self):
        print("Account Details : ")
        print(f"Balance : {self._balance}")
        print(f"Account Number : {self._account_no}\n")
        
    def deposite(self,deposite_money):
        
        if deposite_money > 0 :
            self._balance += deposite_money
            print(f"Money Deposited Successfully. Your Current Balance : {self._balance}")
        
        else :
            print("Invalid input . Please enter the valid input ")
            
    def withdraw(self,withdraw_money):
        
        if withdraw_money <= self._balance:
            self._balance -= withdraw_money
            print(f"Money Withdrawed Successfully. Your Current Balance : {self._balance}")
            
        else :
            print("You don't have enough money")
        

acc1 = BankAccount()     
acc1.account_details()
acc1.deposite(1000)
acc1.withdraw(500)
```

```python
from abc import ABC,abstractmethod

class Employee(ABC):
    def __init__(self, name, role):
        self.name = name
        self.role = role
    
    @abstractmethod
    def display_info(self):
        return f"{self.role}: {self.name}"
    
    @abstractmethod
    def calculate_salary(self):
        raise NotImplementedError("calculate_salary method must be implemented in subclasses")

class Manager(Employee):
    def __init__(self, name, team_size):
        super().__init__(name, role="Manager")
        self.team_size = team_size

    def calculate_salary(self):
        base_salary = 50000
        bonus = self.team_size * 1000
        return base_salary + bonus
    
    def display_info(self):
        return f"{self.role}: {self.name}, Team Size : {self.team_size}"
    

class Developer(Employee):
    def __init__(self, name, programming_language):
        super().__init__(name, role="Developer")
        self.programming_language = programming_language

    def calculate_salary(self):
        base_salary = 60000
        if self.programming_language == "Python":
            bonus = 5000
        else:
            bonus = 3000
        return base_salary + bonus
    
    def display_info(self):
        return f"{self.role}: {self.name} , Programming Language : {self.programming_language}"
    

class Designer(Employee):
    def __init__(self, name, creative_skill_level):
        super().__init__(name, role="Designer")
        self.creative_skill_level = creative_skill_level

    def calculate_salary(self):
        base_salary = 55000
        bonus = self.creative_skill_level * 100
        return base_salary + bonus
    
    def display_info(self):
        return f"{self.role}: {self.name} , Creative Skill Level : {self.creative_skill_level}"
    

    
manager = Manager(name="Harsh", team_size=10)
developer = Developer(name="Jay", programming_language="Python")
designer = Designer(name="Vansh", creative_skill_level=8)

employees = [manager, developer, designer]

for employee in employees:
    print(employee.display_info())
    print(f"Salary: RS {employee.calculate_salary():,.2f}")

```

```python
from abc import ABC , abstractmethod

class System(ABC):
    
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
        self.powered_on = False
     
    @abstractmethod
    def power_on(self):
        pass
    
    @abstractmethod
    def shutdown(self):
        pass
    
    @abstractmethod
    def show_details(self):
        return f"Brand : {self.brand} , Model : {self.model}"
    
class Desktop(System):
    
    def __init__(self,brand,model,processor):
        super().__init__(brand,model)
        self.processor = processor
    
    def power_on(self):
        
        if not self.powered_on :
            
            print(f"{self.brand} {self.model} is powering on !!!")
            self.powered_on = True
        
        else :
            print(f"{self.brand} {self.model} is already  running !!!")
            
    def shutdown(self):
        
        if self.powered_on :
            print(f"{self.brand} {self.model} is shutting down")
            self.powered_on = False
            
        else : 
            print(f"{self.branf} {self.model} is already off")
            
    def show_details(self):
        return f"Brand : {self.brand} , Model : {self.model} , Processor : {self.processor}"
    
        
class Laptop(System):
    
    def __init__(self,brand,model,graphic_card):
        super().__init__(brand,model)
        self.graphic_card = graphic_card
    
    def power_on(self):
        
        if not self.powered_on :
            
            print(f"{self.brand} {self.model} is powering on !!!")
            self.powered_on = True
        
        else :
            print(f"{self.brand} {self.model} is already  running !!!")
            
    def shutdown(self):
        
        if self.powered_on :
            print(f"{self.brand} {self.model} is shutting down")
            self.powered_on = False
            
        else : 
            print(f"{self.brand} {self.model} is already off")
            
    def show_details(self):
        return f"Brand : {self.brand} , Model : {self.model} , Graphics Card : {self.graphic_card}"

dt1 = Desktop("Apple","Desktop","i7")
lp1 = Laptop("Lenovo" ,"Ideapad","RTX Geforce 1650Ti")

dt1.power_on()
dt1.shutdown()
print(dt1.show_details())

print("\n")

lp1.power_on()
lp1.shutdown()
print(lp1.show_details())
```

```python
from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def add_book(self, book):
        pass

    @abstractmethod
    def borrow_book(self, member, book):
        pass

class LibraryBranch(LibrarySystem):
    def __init__(self, name, location):
        super().__init__(name)
        self.location = location
        self.books_available = []

    def add_book(self, book):
        print(f"Adding '{book}' to the {self.name} branch.")
        self.books_available.append(book)

    def borrow_book(self, member, book):
        if book in self.books_available:
            print(f"{member} borrowed '{book}' from the {self.name} branch.")
            self.books_available.remove(book)
        else:
            print(f"Sorry, '{book}' is not available at the moment in the {self.name} branch.")

class Member(LibrarySystem):
    def __init__(self, name):
        super().__init__(name)
        self.books_borrowed = []

    def add_book(self, book):
        print(f"{self.name} added '{book}' to their borrowed books.")
        self.books_borrowed.append(book)

    def borrow_book(self, library_branch, book):
        library_branch.borrow_book(self.name, book)
        self.add_book(book)

library_branch = LibraryBranch(name="City Library", location="Main Street")
member = Member(name="Harsh")

library_branch.add_book("The Great Gatsby")
library_branch.add_book("To Kill a Mockingbird")

member.borrow_book(library_branch, "The Great Gatsby")
member.borrow_book(library_branch, "Harry Potter and the Sorcerer's Stone")

print(f"\nAvailable Books in {library_branch.name}: {library_branch.books_available}")
```

```python
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Now playing: {self.title} by {self.artist} ({self.duration} seconds)")

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []  

    def add_song(self, song):
        self.songs.append(song)

    def play_all(self):
        print(f"Playing all songs in the playlist '{self.name}':")
        for song in self.songs:
            song.play()

class MusicPlayer:
    def __init__(self):
        self.playlists = []  

    def create_playlist(self, name):
        new_playlist = Playlist(name)
        self.playlists.append(new_playlist)
        return new_playlist

    def play_playlist(self, playlist_name):
        for playlist in self.playlists:
            if playlist.name == playlist_name:
                playlist.play_all()
                return
        print(f"Playlist '{playlist_name}' not found.")

song1 = Song(title="Song 1", artist="Artist 1", duration=180)
song2 = Song(title="Song 2", artist="Artist 2", duration=240)
song3 = Song(title="Song 3", artist="Artist 3", duration=200)

playlist1 = Playlist(name="My Favorites")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2 = Playlist(name="Chill Vibes")
playlist2.add_song(song2)
playlist2.add_song(song3)

music_player = MusicPlayer()
music_player.create_playlist("Road Trip")
music_player.create_playlist("Study Mix")
music_player.playlists[0].add_song(song3)

music_player.play_playlist("My Favorites")
music_player.play_playlist("Chill Vibes")
music_player.play_playlist("Road Trip")
music_player.play_playlist("Nonexistent Playlist")
```

```python
class CPU:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def process_data(self):
        print(f"{self.brand} CPU processing data at {self.speed} GHz")

class RAM:
    def __init__(self, capacity):
        self.capacity = capacity

    def store_data(self):
        print(f"RAM storing data. Capacity: {self.capacity} GB")

class Storage:
    def __init__(self, type, capacity):
        self.type = type
        self.capacity = capacity

    def read_data(self):
        print(f"{self.type} storage reading data. Capacity: {self.capacity} GB")

class Computer:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu  
        self.ram = ram  
        self.storage = storage  

    def perform_operation(self):
        print("Computer performing an operation:")
        self.cpu.process_data()
        self.ram.store_data()
        self.storage.read_data()


my_cpu = CPU(brand="Intel", speed=2.5)
my_ram = RAM(capacity=8)
my_storage = Storage(type="SSD", capacity=256)

my_computer = Computer(cpu=my_cpu, ram=my_ram, storage=my_storage)
my_computer.perform_operation()
```

```python
class Engine:
    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

    def start(self):
        print(f"Engine started. Fuel type: {self.fuel_type}")

class Wheels:
    def __init__(self, tire_type):
        self.tire_type = tire_type

    def roll(self):
        print(f"Wheels rolling on {self.tire_type} tires")

class Transmission:
    def __init__(self, transmission_type):
        self.transmission_type = transmission_type

    def shift_gear(self):
        print(f"Transmission shifted to {self.transmission_type}")

class Car:
    def __init__(self, make, model, fuel_type, tire_type, transmission_type):
        self.make = make
        self.model = model
        self.engine = Engine(fuel_type)  
        self.wheels = Wheels(tire_type)  
        self.transmission = Transmission(transmission_type)  

    def start(self):
        print(f"Starting the {self.make} {self.model}")
        self.engine.start()
        self.transmission.shift_gear()
        self.wheels.roll()

my_car = Car(make="Toyota", model="Camry", fuel_type="Gasoline", tire_type="All-season", transmission_type="Automatic")
my_car.start()
```

```python
class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        print(f"Attacking with {self.name}. Damage: {self.damage}")

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def absorb_damage(self, incoming_damage):
        absorbed_damage = min(incoming_damage, self.defense)
        print(f"Absorbing {absorbed_damage} damage with {self.name} armor.")
        return incoming_damage - absorbed_damage

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"Added {item.name} to inventory.")

class GameCharacter:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.weapon = None 
        self.armor = None  
        self.inventory = Inventory() 

    def equip_weapon(self, weapon):
        self.weapon = weapon
        print(f"Equipped {weapon.name}.")

    def equip_armor(self, armor):
        self.armor = armor
        print(f"Equipped {armor.name}.")

    def take_damage(self, incoming_damage):
        if self.armor:
            incoming_damage = self.armor.absorb_damage(incoming_damage)

        self.health -= incoming_damage
        print(f"{self.name} took {incoming_damage} damage. Health: {self.health}")

    def use_inventory_item(self, item_name):
        for item in self.inventory.items:
            if item.name == item_name:
                print(f"{self.name} used {item.name} from the inventory.")
                self.inventory.items.remove(item)
                return
        print(f"{self.name} does not have {item_name} in the inventory.")


sword = Weapon(name="Sword", damage=10)
shield = Armor(name="Shield", defense=5)

player_character = GameCharacter(name="Hero", health=100)
player_character.equip_weapon(sword)
player_character.equip_armor(shield)

enemy_character = GameCharacter(name="Monster", health=50)

player_character.take_damage(8)
player_character.take_damage(15)

health_potion = Weapon(name="Health Potion", damage=0)
player_character.inventory.add_item(health_potion)
player_character.use_inventory_item("Health Potion")
```

```python
class Furniture:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Furniture: {self.name}")

class Appliance:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Appliance: {self.name}")

class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []  
        self.appliances = []  

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display_info(self):
        print(f"\nRoom: {self.name}")
        print("Furniture:")
        for piece in self.furniture:
            piece.display_info()
        print("Appliances:")
        for appliance in self.appliances:
            appliance.display_info()

class House:
    def __init__(self, name):
        self.name = name
        self.rooms = []  

    def add_room(self, room):
        self.rooms.append(room)

    def display_info(self):
        print(f"House: {self.name}")
        for room in self.rooms:
            room.display_info()

living_room = Room(name="Living Room")
couch = Furniture(name="Sofa")
tv = Appliance(name="Television")
living_room.add_furniture(couch)
living_room.add_appliance(tv)

kitchen = Room(name="Kitchen")
table = Furniture(name="Dining Table")
oven = Appliance(name="Oven")
kitchen.add_furniture(table)
kitchen.add_appliance(oven)

my_house = House(name="My Dream House")
my_house.add_room(living_room)
my_house.add_room(kitchen)

my_house.display_info()
```
